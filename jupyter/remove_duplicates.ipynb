```python
import os
import pandas as pd
from IPython.display import display

ROOT_DIR = os.environ.get('PYTHONPATH')

FILENAME = 'vep_benign_variants_with_clusters.csv'
OUTNAME = FILENAME.replace('.csv','') + '_nr' + '.csv'

INFILE_PATH = os.path.join(ROOT_DIR, 'synced_files/vep_variants/', FILENAME)
OUTFILE_PATH = os.path.join(ROOT_DIR, 'synced_files/vep_variants/', OUTNAME)

INFILE_DF = pd.read_csv(INFILE_PATH, sep='\t')
display(INFILE_DF)
```

```python
encountered_dict = {}

def generate_key(tup):
   return getattr(tup, 'uniprot') + str(getattr(tup, 'position')) + getattr(tup, 'WT') + getattr(tup, 'Mut')

boolMask = []

for row in INFILE_DF.itertuples():
   key = generate_key(row)
   try:
      if encountered_dict[key] == True:
         boolMask.append(False)
   except KeyError:
      encountered_dict[key] = True
      boolMask.append(True)

NR_INFILE_DF = INFILE_DF[boolMask]
NR_INFILE_DF['category'] = 'LB/B'

display(NR_INFILE_DF)

NR_INFILE_DF.to_csv(OUTFILE_PATH, sep='\t', index=False)
```
